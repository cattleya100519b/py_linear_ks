# 第6章　対称群による行列式表示・終結式

## 6.2　Pythonで置換を計算

- List 6.1 Permutation.py

```python
import sympy.combinatorics as symc
```

```python
# 恒等置換 (identity permutation)
print(I := symc.Permutation(size=10))

print(" ".join(f"{i:2}" for i in range(len(I.array_form))))
print(" ".join(f"{v:2}" for v in I.array_form))
```

```python
# 巡回置換分解 (cycle decomposition)
print(P := symc.Permutation([2,5,7,3,6,0,1,4,8]))

# 2桁分の幅を確保して縦に並べる場合
print(" ".join(f"{i:2}" for i in range(len(P.array_form))))
print(" ".join(f"{v:2}" for v in P.array_form))
```

> 通常、動かない数字（(3) や (8)）は省略されますが、  
> その置換全体のサイズ（どこまでの数字が含まれているか）を出力から判別できるようにするため、  
> 最大の要素（ここでは 8）の固定点だけは省略されずに (8) として残ります。

```python
# 互いに素 (disjoint) な巡回置換 (cyclic permutation) の積
print(Q := symc.Permutation(1, 2, 3)(5, 4, 6, 9))

print(" ".join(f"{i:2}" for i in range(len(Q.array_form))))
print(" ".join(f"{v:2}" for v in Q.array_form))
```

```python
print(R := symc.Permutation(3, 4)(0, 2, 7))

print(" ".join(f"{i:2}" for i in range(len(R.array_form))))
print(" ".join(f"{v:2}" for v in R.array_form))
```

```python
# 逆置換 (inverse permutation)
print(invQ := Q**-1)

print(f'Q   : {" ".join(f"{v:2}" for v in Q.array_form)}')
print(f'I   : {" ".join(f"{v:2}" for v in I.array_form)}')
print(f'invQ: {" ".join(f"{v:2}" for v in invQ.array_form)}')

print(Q*invQ)
```
