## 3 逆行列

### 3.4 Pythonによる逆行列の求め方

- List 3.1 MatrixInverseCheck.py

（Numpyで）各行列の逆行列を求める。
$$
\begin{pmatrix}
1 & 1 & 0\\
-2 & 0 & 1\\
-1 & 2 & 2\\
\end{pmatrix},\;
\begin{pmatrix}
1 & 1 & 0\\
-2 & 0 & 1\\
0 & 0 & 0\\
\end{pmatrix},\;
\begin{pmatrix}
1 & 2 & 1\\
2 & -1 & 3\\
3 & 1 & 4\\
\end{pmatrix}
$$
（※3つ目の行列はランク2だが、計算誤差で逆行列が算出されてしまう。）

```python
import numpy as np
import numpy.linalg as LA

from common import disp_mat
```

```python
set_A = [np.array([[1,1,0],[-2,0,1],[-1,2,2]]),
    np.array([[1,1,0],[-2,0,1],[0,0,0]]),
    np.array([[1,2,1],[2,-1,3],[3,1,4]])]
```

```python
for i, A in enumerate(set_A, 1):
    print(f'[{i}]')
    disp_mat('A', A)
    print(f'rank(A) = {LA.matrix_rank(A)}')
    try:
        invA = LA.inv(A)
        disp_mat('Inv(A)', invA)
        disp_mat(r'A\,Inv(A)', A@invA)
    except Exception as message:
        print(message)
```

- List 3.2 SympyMatrix.py

Sympy を使えば、誤差なく計算できる。

```python
import sympy
```

```python
for i, A in enumerate(set_A, 1):
    print(f'[{i}]')
    disp_mat('A', A)
    print(f'rank(A) = {LA.matrix_rank(A)}')
    A = sympy.Matrix(A)
    try:
        invA = A.inv()
        disp_mat('Inv(A)', invA)
        disp_mat(r'A\,Inv(A)', A@invA)
    except Exception as message:
        print(message)
```
